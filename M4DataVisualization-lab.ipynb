# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt
#Get the table
QUERY = """
SELECT ConvertedComp
FROM master
"""
ConvComp= pd.read_sql_query(QUERY,conn)

#Create the plot
plt.hist(ConvComp,bins = 15,color='blue', edgecolor='black')

plt.ylabel('Number')
plt.xlabel('Wage')
plt.title('Converted Comp')

plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
#Get the table
QUERY = """
SELECT Age
FROM master
"""
Age= pd.read_sql_query(QUERY,conn)
Age
# Plotting the boxplot
Age.boxplot()

# Adding labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Boxplot of Age')

# Display the plot
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
#Get the table
QUERY = """
SELECT Age,WorkWeekHrs
FROM master
"""
df= pd.read_sql_query(QUERY,conn)

plt.scatter(df['Age'],df['WorkWeekHrs'])

plt.xlabel('Age')
plt.ylabel('Wekkly Hours')
plt.title('Scatter for age and hours')

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
#Get the table
QUERY = """
SELECT Age,WorkWeekHrs,CodeRevHrs
FROM master
"""
df= pd.read_sql_query(QUERY,conn)

# Create the bubble plot
plt.scatter(df['WorkWeekHrs'],df['CodeRevHrs'], s=df['Age'])

# Add labels and title
plt.xlabel('Work Week Hours')
plt.ylabel('Code Rev Hours')
plt.title('Bubble Plot')

# Show the plot
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
#Get the table
QUERY = """
SELECT *
FROM LanguageDesireNextYear
"""
df= pd.read_sql_query(QUERY,conn)

# Filter the data to include only the top 5 databases
top_5_databases = df['LanguageDesireNextYear'].value_counts().head(10)

# Create the pie chart
plt.pie(top_5_databases, labels=top_5_databases.index, autopct='%1.1f%%', startangle=140)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Add a title
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')

# Show the plot
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT Age,WorkWeekHrs,CodeRevHrs
FROM master
"""
df= pd.read_sql_query(QUERY,conn)

# Filter data for respondents aged 30 to 35
filtered_df = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Calculate median WorkWeekHrs and CodeRevHrs
median_values = filtered_df.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()

# Plot stacked chart
median_values.plot(kind='bar', stacked=True)

# Add labels and title
plt.xlabel('Age Group')
plt.ylabel('Median Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')

# Show plot
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
# your code goes here
QUERY = """
SELECT Age,ConvertedComp
FROM master
"""
df= pd.read_sql_query(QUERY,conn)

# Filter data for respondents aged 30 to 35
filtered_df = df[(df['Age'] >= 25) & (df['Age'] <= 30)]

# Create the line chart
plt.plot(filtered_df['Age'],filtered_df['ConvertedComp'])

# Add labels and title
plt.xlabel('Converted Comp')
plt.ylabel('Age')
plt.title('Line Chart')

# Show the plot
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch
FROM master
"""
df= pd.read_sql_query(QUERY,conn)
df.head()

# Count the occurrences of each unique value in the MainBranch column
branch_counts = df['MainBranch'].value_counts()

# Plot a horizontal bar chart
branch_counts.plot(kind='barh')

# Add labels and title
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.title('Main Branch Distribution')

# Show the plot
plt.show()
```
